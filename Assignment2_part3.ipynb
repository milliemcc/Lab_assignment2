```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#import dataset
df = pd.read_csv('/arcgis/home/portland_park_trees.csv')
```

#### Create a scatterplot for just trees in the Ulmus genus with DBH mapped to the x-axis and tree height mapped to the y-axis

```python
ulmus = df.query('Genus=="Ulmus"') #create a subplot for just Ulmus genus

#set the x and y axis
x = ulmus["DBH"]
y = ulmus["TreeHeight"]

#generate the plot
plt.scatter(x, y)
#set labels
plt.title("Ulmus DBH vs tree height")
plt.xlabel("DBH")
plt.ylabel("Tree height")
```

#### Graph 2: Create a scatterplot for just trees in the Ulmus genus with DBH mapped to the x-axis, tree height mapped to the y-axis, and tree species mapped to hue

```python
import seaborn as sns #import seaborn for this graph

ulmus = df.query('Genus=="Ulmus"') #create a subplot for the Ulmus genus

sns.relplot(x="DBH", y="TreeHeight", hue="Species", data=ulmus) 
#create a scatterplot using seaborn where x=DBH, y=Tree height and hue is based on different species in the ulmus subplot

plt.title('Scatterplot of Ulmus species')
```

#### Graph 3: Create a boxplot of DBH for just the Ulmus genus differentiated by species 

```python
#use seaborn for this graph
ulmus = df.query('Genus=="Ulmus"') #create a subplot for the ulmus genra

sns.boxplot(data=ulmus, x="Species", y="DBH") #plot the boxplot of DBH by species
plt.title('Boxplot of DBH') #assign a title
```

#### Graph 4:  Combine Graphs 1 and 3 into a single figure. Do not plot a legend for any of the graphs.

```python
ulmus = df.query('Genus=="Ulmus"') #create a subplot of only Ulmus genre

#set x and y for ulmus
x = ulmus["DBH"]
y = ulmus["TreeHeight"]

fig, (ax1, ax2) = plt.subplots(1,2) #define a figure where there are two different subplots
ax1.scatter(x, y) #the subplot on the first axis is for graph 1 - the scatterplot using matplotlib
sns.boxplot(data=ulmus, x="Species", y="DBH", ax=ax2) #the subplot on the second axis is for graph 2 - a boxplot created using seaborn
```
